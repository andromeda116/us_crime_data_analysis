### Weighted Average Calculator

```python
# Create weighted averages
def create_weighted_avgs(df, avg_name, weight_name):
    print(list(df))
    d = df[avg_name]
    w = df[weight_name]
    try:
        return (d * w).sum() / w.sum()
    except ZeroDivisionError:
        return d.mean()
```

###  3. violent_crime counts for final main L,M,S and final main without outliers L,M,S over years
### (separate agency level comparison graphs) 

```python
#### Utility function to graph counts and rates at agency levels ####
def graph_trends(rows, cols, req_dfs, var, titles, x_label, y_label, weights=None, pop_var=None):
    fig, axs = plt.subplots(nrows=rows, ncols=cols, sharex='col', figsize=(12,5))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    count = 0
    # axes are in a two-dimensional array, indexed by [row, col]
    for i in range(rows):
        for j in range(cols):
            if weights:
                axs[i, j].plot(Years, req_dfs[count].groupby("YEAR").apply(create_weighted_avgs, f'{var}', f'{pop_var}'))
            else:
                axs[i, j].plot(Years, req_dfs[count].groupby('YEAR')[f'{var}'].mean())
            
            axs[i, j].set_title(f'{titles[count]}')

            count += 1

    for ax in axs.flat:
        ax.set(xlabel=f'{x_label}', ylabel=f'{y_label}')

    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axs.flat:
        ax.label_outer()


    fig.tight_layout()
```

```python
graph_trends(rows=2, 
             cols=3, 
             req_dfs = [fnl_mn_large, fnl_mn_med, fnl_mn_sml, fnl_mn_wdt_low_3z_large, fnl_mn_wdt_low_3z_med, fnl_mn_wdt_low_3z_sml],
             var='violent_crime',
             titles = ['Large Agencies', 'Medium Agencies', 'Small Agencies', 'Large Agencies(No Low 3Z)', 
              'Medium Agencies(No Low 3Z)', 'Medium Agencies(No Low 3Z)'],
             x_label = 'Years',
            y_label='Violent Crime Counts')
```
