### Perform WOLS

```python
def wols_multiple_data_multiple_ind_vars(req_dfs, dep_vars, ind_vars, weights_var, dataset_names=False, counts=True):
    df_count = 0
    
    for df in req_dfs:
        for ind_var in ind_vars[df_count]:
            model = sm.WLS.from_formula(f'{dep_vars[df_count]} ~ {ind_var}', data=df, weights=df[f'{weights_var}'])
            results = model.fit()
            if dataset_names:
                print(dataset_names[df_count])
            print(results.summary(), '\n', '\n')           
                
        df_count += 1
```

### Create a new weight col with desired weighting technique

```python
def create_wts_col(dfs, to_wt_col, wt_type):
    for df in dfs:
        if wt_type=='log':
            df[f'{to_wt_col}_{wt_type}'] = np.log(df[f'{to_wt_col}'])
        elif wt_type=='sq_root':
            df[f'{to_wt_col}_{wt_type}'] = np.sqrt(df[f'{to_wt_col}'])
```

### Weights - Inverse of the absolute fitted values obtained from the OLS model

```python
def wols_inv_fitted_values(req_dfs, dep_vars, ind_vars, dataset_names=False, counts=True):
    df_count = 0
    
    for df in req_dfs:
        for ind_var in ind_vars[df_count]:
            model = sm.WLS.from_formula(f'{dep_vars[df_count]} ~ {ind_var}', data=df)
            results = model.fit()
            
            # fitted values (need a constant term for intercept)
            model_fitted_y = results.fittedvalues
            
            # model residuals
            model_residuals = results.resid

            # absolute residuals
            model_abs_resid = np.abs(model_residuals)
            
            ## Compute the weight and add it to the column named "weight"
            df["weight"] = np.abs(model_fitted_y)  
            df["weight"] = df["weight"]**-1
            

            ## add a intercept point
            # x = sm.add_constant(x)
            df[f'{dep_vars[df_count]}'] = sm.add_constant(df[f'{dep_vars[df_count]}']) 

            ## Compute the weight and add it to the column named "weight_1"
            df["weight_1"] = model_fitted_y 
            
            model1 = sm.WLS.from_formula(f'{dep_vars[df_count]} ~ {ind_var}', weights=df['weight'],
                                        data=df) 
            results1 = model1.fit() 
            if dataset_names:
                print(dataset_names[df_count])
            print(results1.summary(), '\n', '\n')         
                
        df_count += 1
```
